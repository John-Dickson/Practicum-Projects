## introduction

The project in front of us today is to mode and predict the gold extraction from ore. 

We have 3 datasets to start with, a training set, a test set, and a full set. The training set is missing some columns as that data was not collected when that dataset was created. The full dataset was created later and contains these missing columns, so we can fill them from there. 

We will first check the data, find any missing values or outliers and figure out how to fill that data, we will then analize the data, try to identify patterns and remove any outliers identified. 

Finally we will create our models, we will use 2 different models and calculate the sMAPE values for them. There are 2 output columns in this dataset, we will calculate them together. Finally we will do cross validation to make sure our models are reliable.

##### rougher.output.recovery formula

'rougher.output.recovery'= ('rougher.output.concentrate_au' * ('rougher.input.feed_au' - 'rougher.output.tail_au')) / ('rougher.input.feed_au' * ('rougher.output.concentrate_au' - 'rougher.output.tail_au'))

```python
C = df_train['rougher.output.concentrate_au']
F = df_train['rougher.input.feed_au']
T = df_train['rougher.output.tail_au']
def recov_au():
    rough_calc = pd.Series((C * (F - T)) / (F * (C - T)) * 100, index=rough_train.index)
    return rough_calc
```

##### Calculate the accuracy of rougher.output.recovery

```python
rough_train = df_train['rougher.output.recovery']
```

```python
print(rough_train.isna().sum())
rough_train.dropna(inplace=True)
rough_train.isna().sum()
```

```python
rough_calc = recov_au()
```

```python
print(rough_calc.isna().sum())
print(rough_calc.isna().sum() / df_train.shape[0] * 100)
rough_calc.dropna(inplace=True)
rough_calc.isna().sum()
```

```python
print("%.20f" % float(mean_absolute_error(rough_calc, rough_train)))
```

### Filling missing values

#### function for missing values

```python
#This function calculates the percent of missing values in any column
def mis_val(df):
    for col in df.columns.tolist():
        missing_val = df[col].isna().sum() / df.shape[0] * 100
        if missing_val > 0:
            print(col, 'NaN%:', missing_val)
    print('done')
```

We created a function that sets the index to be the date column, and then use interpolate to fill missing values. We were able to run the interpolate function with both `ffill` and `linear` to determine that linear was the best way to fill the missing values based on the models. 

```python
def lin_fill(df):
    df.set_index('date', inplace=True)
    df.interpolate(method ='linear', limit_direction ='forward', inplace=True)
    mis_val(df_train)
    return df
```

### Analysis

#### Function for graphing

We create a somewhat complex function for graphing, this allows us to graph each of the values we want with a single line of code, and allows us to come back and add or remove things in seconds. 

```python
mat_name = ['ag', 'pb', 'sol', 'au']
def met_plot(hist_rows, mat_name, graph_name, plot_rng):
    print(graph_name)
    plt.figure(figsize=(20,9))
    for i in range(len(hist_rows)):
        plt.hist(features_train[hist_rows[i]], bins=30, alpha=0.5, label=mat_name[i], range=plot_rng)
#     plt.figure(figsize=(9,20))
    plt.legend(loc='best')
    plt.show()
    fig, ax = plt.subplots()
    ax.boxplot((
        features_train[hist_rows[0]],
        features_train[hist_rows[1]], 
        features_train[hist_rows[2]], 
        features_train[hist_rows[3]]))
    ax.set_xticklabels(mat_name)
    ax.set_xlabel("Metal")
    ax.set_ylabel("output")
    ax.set_title(graph_name)
    fig.set_figwidth(20)
    fig.set_figheight(9)
    plt.show()
```

We create a function for dropping the values that we don't want, again saving time and avoiding us having to type the same thing too many times. 

```python
def drop_low_val(col):
    for c in col:
        features_train.drop(features_train[features_train[c] <= 0].index, inplace=True)
        features_test.drop(features_test[features_test[c] <= 0].index, inplace=True)
```

#### Linear Regression model

```python
def smape(a, f):
    return 1/len(a) * np.sum(2 * np.abs(f-a) / (np.abs(a) + np.abs(f))*100)
```

```python
def gold_model(features_train, features_valid, target_train, target_valid):
    model = LinearRegression()
    model.fit(features_train, target_train) # train model on training set
    predictions_valid = model.predict(features_valid) # get model predictions on validation set
    print('real means:', target_valid.mean())
    predictions_valid = pd.DataFrame(predictions_valid, columns=('rougher.output.recovery', 'final.output.recovery'), index=target_test.index)
    print('predicted means:', predictions_valid.mean())
    result = mean_squared_error(target_valid, predictions_valid)**0.5
    print("RMSE of the linear regression model on the validation set:", result)
    f_smape = smape(target_test, predictions_valid)
    print("sMAPE of the linear regression model on the test set:\n", f_smape)
    print("final sMAPE:", (0.25*f_smape[0]) + (0.75*f_smape[1]))
    return predictions_valid
```

## Conclusions

From this we have created an accurate model of our output recovery for both rougher and final stages. There are interestingly 2 sets of paramaters that we discovered that give exactly the same sNAPE value (depth=30, n_estimators 150) and (depth=17, n_estimators=300). The best option is the one with depth 30, both because of it's faster runtime and because it gives marginally better cross validation score. 

The model that should be chosen for predicting future output is `RandomForestRegressor` with `depth=30` and `n_estimators=150`. This will give the most accurate results with the lowest processing time. The models tend to overfit, perhaps with some more tuning this could be resolved. 

With 86 columns of data to work with we were not able to assess every individual column and remove outliers, that would take days to do by itself. If this were done it might result in marginally more accurate models, perhaps even increasing the sNAPE value by another 0.5% or so.
